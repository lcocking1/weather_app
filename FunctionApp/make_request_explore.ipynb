```python
forecast_items = forecast_container_client.query_items(
    "select * from c where not is_defined(c.gridId)",
    enable_cross_partition_query=True
)
```

```python
id = test_item["id"]
id_split = id.split('_')
grid_id = id_split[0]
grid_x = id_split[1]
grid_y = id_split[2]
id_ts = id_split[3]
```

```python
# update forecast_items to have relationship to gridpoints
for item in forecast_items:
    id = item["id"]
    id_split = id.split('_')
    grid_id = id_split[0]
    grid_x = id_split[1]
    grid_y = id_split[2]
    id_ts = id_split[3]
    update_dict = item
    update_dict["gridId"] = grid_id
    update_dict["gridX"] = grid_x
    update_dict["gridY"] = grid_y
    update_dict["forecast_generated_timestamp"] = id_ts
    forecast_container_client.replace_item(id, update_dict)
```

```python
# replace gridpoints with coordinates 
for gridpoint in gridpoints_generator:
    temp_dict = {}
    temp_dict["id"] = gridpoint["id"]
    temp_dict["gridId"] = gridpoint["gridId"]
    temp_dict["gridX"] = gridpoint["gridX"]
    temp_dict["gridY"] = gridpoint["gridY"]
    temp_dict["coordinates"] = [gwd.search_coordinates(i["coordinates"]) for i in forecast_items_dict if 
                     i["gridId"] == gridpoint["gridId"] 
                     and i["gridX"] == gridpoint["gridX"] 
                     and i["gridY"] == gridpoint["gridY"]]
    gridpoints_container_client.replace_item(
        item=gridpoint["id"],
        body=temp_dict
    )
```
